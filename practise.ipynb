```python
class Student:
    def __init__(self,name,marks):
       self.name =name
       self.marks =marks
    def display(self):
        print(self.name,self.marks)
s1 = Student("Sneha",80)
s1.display()
```

```python
class Book:
    def __init__(self, title):
        self.title =title
b1 = Book("python Basics")
print(b1.title)
```

```python
class parent:
    def fun(self):
        print("parent method")
class child(parent):
    pass
c = child()
c .fun()
```

```python
class Father:
    def skill1(self):
        print("Coding")
class Mother:
    def skill2(self):
        print("singing")
class child(Father,Mother):
    pass
c = child()
c.skill1()
c.skill2()
```

```python
class Animal:
    def sound(self):
        print("Animal sound")
class Dog(Animal):
    def sound(self):
        print("Bark")
d = Dog()
d.sound()
```
