```python
import pandas as pd
data={"Name":["mani", "mamu", "baby", "sanju"],
      "Age":[33,40,51,20],
      "City":["knr","hyd","pdpl","knr"]
     }
details=pd.DataFrame(data)
print(details[details.Age>31]),(details[details.Age<29])
details["Grade"]=["A","B","C","D"]
print(details)
details["Age"]=details["Age"]+3
print(details)
```
