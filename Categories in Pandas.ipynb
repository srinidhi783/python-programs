```python
import matplotlib.pyplot as plt
products=["mobile", "laptop", "tablet"]
sales=[500,300,200]
plt.bar(products,sales)
plt.show()
plt.barh(products,sales)
plt.show()
```

```python
import numpy as np
products=["mobile","laptop","tablet"]
sales_2024=[400,500,600]
sales_2025=[800,900,1000]
x=np.arange(len(products))
width=0.35
plt.bar(x-width/2,sales_2024,width)
plt.bar(x-width/2,sales_2025,width)
plt.show()
```

```python
import numpy as np
products=["mobile","laptop","tablet","tv"]
offline=[200,300,400,500]
online=[600,70,800,900]
plt.bar(products,online)
plt.bar(products,offline,bottom=online)
plt.show()
```
