```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\burra\OneDrive\Documents\students.csv")
```

```python
df
```

```python
df["Total"] = df["Maths"] + df["Physics"] + df["Computer"]
df
```

```python
df["Average"] = df["Total"] / 3
df
```

```python
print("Mean:", df["Average"].mean())
print("Median:", df["Average"].median())
print("Mode:", df["Average"].mode()[0])
```

```python
import matplotlib.pyplot as plt
plt.bar(df["Name"], df["Average"])
plt.title("Student Average Marks")
plt.xlabel("Students")
plt.ylabel("Students")
plt.show()
```

```python
df["Grade"] = "Pass"
grade_count = df["Grade"].value_counts()
plt.pie(grade_count,
labels=grade_count.index, autopct="%1.1f%%")
plt.show()
```

```python
plt.hist(df["Average"])
plt.title("Average Marks Distribution")
plt.show()
```

```python
class Student:
    def __init__(self, name, maths, physics, computer):
        self.name = name
        self.maths = maths
        self.physics = physics
        self.computer = computer

    def total(self):
        return self.maths + self.physics + self.computer
```

```python
df.to_csv("student_report.csv", index=False)
```
