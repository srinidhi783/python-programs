```python
import pandas as pd
employee={
    "name":["arun", "pinky", "rose", "phani", "indhu"],
    "salary":[20000,25000,30000,35000,40000]
}
result=pd.DataFrame(employee)
print(result)
print(result[result.salary>23000])
print(result.query("salary>23000"))
print(result.query("salary>23000&salary<35000"))
```

```python
import pandas as pd
data={
    "Date":["2026-1-8", "2025-2-8", "2024-3-8", "2023-4-8", "2022-5-8"]
}
result=pd.DataFrame(data)
print(result)
result["Date"]=pd.to_datetime(result["Date"])
print(result)
print(result["Date"].dt.year)
print(result["Date"].dt.month)
print(result["Date"].dt.day)
pd.Timestamp.now()
```
