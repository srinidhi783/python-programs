```python
import matplotlib.pyplot as plt
products=["mobiles", "laptop", "tablet"]
Sales=[800,400,500]
plt.barh(products,Sales)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
products=["mobile", "laptop", "tv"]
sales_2024=[400,500,600]
sales_2025=[600,700,800]
x=np.array(len(products))
width=0.35
plt.bar(x-width/2,sales_2024,width,label='2024')
plt.bar(x-width/2,sales_2025,width,label='2025')
plt.xlabel("products")
plt.title("sales comparison: 2024 vs 2025")
plt.legend()
plt.show()
```

```python
import numpy as np
products=["mobile", "laptop", "tv"]
online=[1000,2000,3000]
offline=[1000,2300,3000]
wholesale=[1500,3000,3500]
plt.bar(products,online,label='online')
plt.bar(products,offline,label='offline')
plt.bar(products,wholesale,label='wholesale')
bottom_value=[online[i]+offline[i]for i in range(len(products))]
plt.bar(products,wholesale,bottom=bottom_value,label='wholesale')
plt.xlabel("products")
plt.title("Revenue of the Distributed Sales")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may"]
sales=[100,200,300,400,500]
sns.lineplot(x=months,y=sales)
plt.xlabel("products")
plt.ylabel("sales")
plt.title("monthly sales trends")
plt.show()
```

```python
import plotly.express as px
import pandas as pd
df=pd.DataFrame({
    'months':['jan','feb','mar','apr','may'],
    'sales':[100,200,300,400,500]})
fig=px.line(df,x='months',y='sales' ,markers=True)
fig.show()
```
